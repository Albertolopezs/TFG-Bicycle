```python
def run_simulation(workers_number, temperature_scenario, automation_scenario, workload_scenario):
    stress_model = StressModel(workers_number, temperature_scenario, automation_scenario, workload_scenario)

    for i in range(43200):
        stress_model.step()
    
    workers_data = stress_model.worker_collector.get_agent_vars_dataframe()
    time_data = stress_model.time_collector.get_agent_vars_dataframe()
    model_data = stress_model.model_collector.get_model_vars_dataframe()
    workers_simulations_data.append(workers_data)
    time_simulations_data.append(time_data)
    model_simulations_data.append(model_data)
```

```python
def run_simulation_once(workers_number, temperature_scenario, automation_scenario, workload_scenario):
    stress_model = StressModel(workers_number, temperature_scenario, automation_scenario, workload_scenario)

    for i in range(86400):
        stress_model.step()
    
    workers_data = stress_model.worker_collector.get_agent_vars_dataframe()
    time_data = stress_model.time_collector.get_agent_vars_dataframe()
    model_data = stress_model.model_collector.get_model_vars_dataframe()

    return workers_data, time_data, model_data
```

```python
# For a jupyter notebook add the following line:
%matplotlib inline
import numpy as np

def draw_average_stress(model_data, time_data, automation):
    model_data[['Average Stress','Average Productivity']].plot(linewidth=2)
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    plt.xticks(range(0, len(time_data.Day.index)), 
               ["Day {} - {}".format(b_, a_) for a_, b_ in zip(time_data.Hour, time_data.Day)], 
               size='small', rotation=90)
    axes = plt.gca()
    axes.set_ylim([0, 1])
    xticks = axes.xaxis.get_major_ticks()
    i = 0
    
    plt.xlabel('Time', size=15)
    plt.ylabel('Level', size=15)
    plt.title('Stress and productivity (' + automation + ')', size=20)
    
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='on') # labels along the bottom edge are off
    
    for tick in axes.yaxis.get_major_ticks():
        tick.label.set_fontsize(14)
        
    for xtick in xticks:
        if i % 4 != 0:
            xtick.label1.set_visible(False)
        else:
            xtick.label1.set_fontsize(14) 
        i += 1

def draw_average_stress_components(model_data, time_data, automation):
    model_data[['Average Event Stress','Average Effective Fatigue', 'Average Time Pressure']].plot(linewidth=2)
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    plt.xticks(range(0, len(time_data.Day.index)), 
               ["Day {} - {}".format(b_, a_) for a_, b_ in zip(time_data.Hour, time_data.Day)], 
               size='small', rotation=90)
    axes = plt.gca()
    axes.set_ylim([0, 1])
    xticks = axes.xaxis.get_major_ticks()
    i = 0
    
    plt.xlabel('Time', size=15)
    plt.ylabel('Level', size=15)
    plt.title('Stress and productivity (' + automation + ')', size=20)
    
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='on') # labels along the bottom edge are off
    
    for tick in axes.yaxis.get_major_ticks():
        tick.label.set_fontsize(14)
        
    for xtick in xticks:
        if i % 4 != 0:
            xtick.label1.set_visible(False)
        else:
            xtick.label1.set_fontsize(14) 
        i += 1

def draw_average_stress_components_histogram(baseline_workers_data, ambient_workers_data, workload_workers_data, full_workers_data):
   # data to plot
    n_groups = 4
    means_frank = (baseline_workers_data['Stress'].mean(), ambient_workers_data['Stress'].mean(), workload_workers_data['Stress'].mean(), full_workers_data['Stress'].mean())
    means_guido = (baseline_workers_data['Event stress'].mean(), ambient_workers_data['Event stress'].mean(), workload_workers_data['Event stress'].mean(), full_workers_data['Event stress'].mean())
    means_frank_1 = (baseline_workers_data['Time pressure'].mean(), ambient_workers_data['Time pressure'].mean(), workload_workers_data['Time pressure'].mean(), full_workers_data['Time pressure'].mean())
    means_guido_1 = (baseline_workers_data['Effective fatigue'].mean(), ambient_workers_data['Effective fatigue'].mean(), workload_workers_data['Effective fatigue'].mean(), full_workers_data['Effective fatigue'].mean())
    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.15
    opacity = 0.8

    rects1 = plt.bar(index+bar_width, means_frank, bar_width,
                     alpha=opacity,
                     color='b',
                     label='Stress')

    rects2 = plt.bar(index+bar_width*2, means_guido, bar_width,
                     alpha=opacity,
                     color='g',
                     label='Event Stress')
    rects3 = plt.bar(index+bar_width*3, means_frank_1, bar_width,
                     alpha=opacity,
                     color='r',
                     label='Time Pressure')

    rects4 = plt.bar(index+bar_width*4, means_guido_1, bar_width,
                     alpha=opacity,
                     color='y',
                     label='Effective Fatigue')

    plt.xlabel('Policy', size=15)
    plt.ylabel('Stress level', size=15)
    plt.title('Stress components by policy', size=20)
    plt.xticks(index + bar_width*3, ('Baseline', 'Ambient Automation', 'Workload Automation', 'Full automation'))
    plt.legend()
    
    
    ax = plt.gca()
    ax.set_ylim([0,1])
    xticks = ax.xaxis.get_major_ticks()
    i = 0

    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(14)

    for xtick in xticks:
        xtick.label1.set_fontsize(14) 
        i += 1
        
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    
    plt.tight_layout()
    plt.show()
    

def draw_average_stress_all_policies(model_data, time_data):
    model_data[['Baseline', 'Ambient', 'Workload', 'Full']].plot(linewidth=2)
    fig = plt.gcf()
    plt.legend(loc=2, prop={'size': 16})
    fig.set_size_inches(15,10)
    plt.xticks(range(0, len(time_data.Day.index)), 
               ["{}".format(b_) for b_ in time_data.Day], 
               size='small', rotation=0)
    axes = plt.gca()
    axes.set_ylim([0, 1])
    xticks = axes.xaxis.get_major_ticks()
    i = 0
    
    plt.xlabel('Day', size=15)
    plt.ylabel('Stress', size=15)
    plt.title('Average stress level', size=20)
    
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='on') # labels along the bottom edge are off
    
    for tick in axes.yaxis.get_major_ticks():
        tick.label.set_fontsize(14)
        
    for xtick in xticks:
        if i % 4 != 0 and i % 59 != 0:
            xtick.label1.set_visible(False)
        else:
            xtick.label1.set_fontsize(14) 
        i += 1

def draw_average_productivity_all_policies(model_data, time_data):
    model_data[['Baseline', 'Ambient', 'Workload', 'Full']].plot(linewidth=2)
    fig = plt.gcf()
    plt.legend(loc=3, prop={'size': 16})
    fig.set_size_inches(15,10)
    plt.xticks(range(0, len(time_data.Day.index)), 
               ["{}".format(b_) for b_ in time_data.Day], 
               size='small', rotation=0)
    axes = plt.gca()
    axes.set_ylim([0, 1])
    xticks = axes.xaxis.get_major_ticks()
    i = 0
    
    plt.xlabel('Day', size=15)
    plt.ylabel('Productivity', size=15)
    plt.title('Average productivity level', size=20)
    
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='on') # labels along the bottom edge are off
    
    for tick in axes.yaxis.get_major_ticks():
        tick.label.set_fontsize(14)
        
    for xtick in xticks:
        if i % 4 != 0 and i % 59 != 0:
            xtick.label1.set_visible(False)
        else:
            xtick.label1.set_fontsize(14) 
        i += 1
```

### Define the problem

```python
from model import BikeModel
def run_simulation(walking_max_dist,riding_max_dist,initial_trucks,percTrips,truck_speed,bike_speed):
    bike_model = BikeModel(1,walking_max_dist,riding_max_dist,initial_trucks,percTrips,truck_speed,bike_speed)

    for i in range(60*24):
        bike_model.step()
    
    #workers_data = stress_model.worker_collector.get_agent_vars_dataframe()
    #time_data = stress_model.time_collector.get_agent_vars_dataframe()
    model_data = bike_model.datacollector.get_model_vars_dataframe()
    #workers_simulations_data.append(workers_data)
    #time_simulations_data.append(time_data)
    model_simulations_data.append(model_data)

```

```python
morris_problem = {
    # There are six variables
    'num_vars': 6,
    # These are their names
    'names': [  'bike_speed',
                'truck_speed',
                'truck_number',
                'walking_max',
                'riding_max',
                'number_trips'
             ],
    # Plausible ranges over which we'll move the variables
    'bounds': [[10,15],  # bike_speed
               [30,50],  # truck_speed
               [1,10],   # truck_number
               [250,750],   # walking_max
               [375,1125],   # riding_max
               [0.01, 1]   # numbertrips
              ],
    # I don't want to group any of these variables together
    'groups': None
    }
```

### Generate a sample

```python
number_of_trajectories = 10
sample = ms.sample(morris_problem, number_of_trajectories, num_levels=5)
```
